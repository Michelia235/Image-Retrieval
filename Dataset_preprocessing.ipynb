## **2. Xử lí dữ liệu - Tổ chức cấu trúc folder**

```python
import os
import shutil
from collections import defaultdict

# Xác định thư mục nguồn và thư mục đích
source_dir = "Dataset"
train_dir = "data/train"
test_dir = "data/test"

# Tạo thư mục đích nếu chưa tồn tại
os.makedirs(train_dir, exist_ok=True)
os.makedirs(test_dir, exist_ok=True)

# Khởi tạo một từ điển để lưu trữ đường dẫn tệp cho mỗi lớp
class_files = defaultdict(list)

# Đọc đường dẫn tệp từ tệp văn bản
with open('filename.txt', 'r') as file:
    lines = file.readlines()
    for line in lines:
        line = line.strip()  # Xóa khoảng trắng đầu và cuối
        if line:
            # Trích xuất tên lớp từ đường dẫn
            parts = line.split('/')
            class_name = parts[2]  # Cấu trúc Dataset/category/class/image.jpg
            class_files[class_name].append(line)

# Di chuyển hình ảnh vào các thư mục train và test
for class_name, files in class_files.items():
    # Tạo thư mục train và test cho lớp
    train_class_dir = os.path.join(train_dir, class_name)
    test_class_dir = os.path.join(test_dir, class_name)
    os.makedirs(train_class_dir, exist_ok=True)
    os.makedirs(test_class_dir, exist_ok=True)

    # Di chuyển 19 hình ảnh vào thư mục train và 1 hình ảnh vào thư mục test
    for i, file_path in enumerate(files):
        if i == 0:
            shutil.copy(file_path, test_class_dir)  # Di chuyển hình ảnh đầu tiên vào thư mục test
        elif i < 20:
            shutil.copy(file_path, train_class_dir)  # Di chuyển 19 hình ảnh còn lại vào thư mục train

print("Tổ chức dữ liệu hoàn tất!")

# Nén thư mục dữ liệu để lưu trữ hoặc chia sẻ
!zip -r /content/drive/MyDrive/data.zip data
```
